### Section 2.9, code block 2.11

```python
# vector
v = np.array([1,2])

# plot it
plt.plot([0,v[0]],[0,v[1]])

# and then plot scaled versions on top
for i in range(10):
    # random scalar
    s = np.random.randn()
    # scaled vector
    sv = s*v
    plt.plot([0,sv[0]],[0,sv[1]])
    
    
plt.grid('on')
plt.axis('square')
plt.axis([-4,4,-4,4])
plt.show()
```
